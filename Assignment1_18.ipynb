```python
#1 Barplot
plt.bar(data['PL'],data['PW'])
plt.title('Petal Length VS Petal Width')
plt.ylabel('Petal Width')
plt.xlabel('Petal Length')
```

```python
#2  Scatterplot of SL vs SW
import seaborn as sns
sns.scatterplot(x='SL', y='SW',hue='Classification', data=data, )
```

```python
#3 scatterplot of PL vs PW
sns.scatterplot(x='PL', y='PW',hue='Classification', data=data, )
```

```python
#4 boxplot
sns.boxplot(x="Classification", y="PL", palette="husl", data=data)
```

```python
#5 boxplot
data.boxplot(by="Classification", figsize=(12, 10))
```

```python
#6 Pairplot
sns.pairplot(data)
```

```python
#7 Heatmap
sns.heatmap(data.corr(), annot=True)
```
