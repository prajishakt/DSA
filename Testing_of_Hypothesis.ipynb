● The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.


Null Hypothesis (H₀): There is no significant increase in sales after implementing digital marketing.

Alternative Hypothesis (H₁): There is a significant increase in sales after implementing digital marketing.


```python
#Perform T-test
from scipy.stats import ttest_ind

t_statistic, p_value = ttest_ind(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
```

```python
print("p value=",p_value)
if p_value<0.05:
  print("Null hypothesis rejected")
else:
  print('Null hypothesis Accepted')
```

There is a significant increase in sales after implementing digital marketing.


2.The company needs to check whether there is any dependency between thefeatures “Region” and “Manager”.

Null Hypothesis (H₀): There is no dependency between Region and Manager.

Alternative Hypothesis (H₁): There is a dependency between Region and Manager.


```python
#Perform chi square test
from scipy.stats import chi2_contingency
#Create a Contingency Table:
table=pd.crosstab(df['Region'],df['Manager'])
table
```

```python
from scipy.stats import chi2_contingency
chi2_stat,p_value2,dof,exp=chi2_contingency(table)
print("p value=",p_value2)
if p_value2<0.05:
  print("Null hypothesis rejected")
else:
  print('Null hypothesis Accepted')
```

There is no dependency between region and manager. These two features are independent.
