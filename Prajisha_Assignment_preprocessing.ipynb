```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

4. Fill in all the missing values present in all the columns in the dataset


```python
data.isna().sum() #Checking null values
```

```python
#Fill columns having null values
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
```

```python
data.isna().sum()
```

5. Check and handle outliers in at least 3 columns in the dataset


```python
data.info()
```

```python
num_columns=['Pclass','Age','SibSp','Parch','Fare']
for i in num_columns:
  plt.figure()
  plt.boxplot(data[i])
  plt.title(i)
```

```python
#Handling outliers with IQR method
cols = ['Pclass','SibSp','Parch','Fare']
for i in cols:
  Q1=np.percentile(data[i],25,method='midpoint')
  Q2=np.percentile(data[i],50,method='midpoint')
  Q3=np.percentile(data[i],75,method='midpoint')
  IQR=Q3-Q1
  low_lim = Q1-1.5*IQR
  up_lim = Q3+1.5*IQR
  outlier=[]
  for x in data[i]:
    if(x<low_lim) or (x>up_lim):
      outlier.append(x)
  print(outlier)
  data[i] = data[i].clip(lower=low_lim,upper=up_lim) #clip method to handle outliers
```

```python
#Handling outlier of Age using zscore method
from scipy.stats import zscore
z_scores = zscore(data['Age'])
print(z_scores)
```

```python
threshold=3
outliers_index=data.index[(z_scores>threshold) | (z_scores<-threshold)]
print(outliers_index)
df1=data.drop(outliers_index)
df1.head()
```

```python
num_columns=['Pclass','Age','SibSp','Parch','Fare']
for i in num_columns:
  plt.figure()
  plt.boxplot(df1[i])
  plt.title(i)
```
